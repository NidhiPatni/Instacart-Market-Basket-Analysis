### Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import dask.dataframe as dd

# code to ignore warnings
import warnings
warnings.filterwarnings("ignore")

# garbage collector to free up memory
import gc
gc.enable()

root_csv = '../csv files/'
root_pickle = '../pickle files/'
```

### Creating a function to reduce the memory of dataframe

```python
def reduce_mem_usage(train_data):
    start_mem = train_data.memory_usage().sum() / 1024**2
    
    for col in train_data.columns:
        col_type = train_data[col].dtype

        if not pd.api.types.is_numeric_dtype(train_data[col]):
            # Handle non-numeric columns (e.g., object, datetime) appropriately
            # You might choose to skip these columns or apply a different logic
            continue
        
        if pd.api.types.is_categorical_dtype(train_data[col]):
            train_data[col] = train_data[col].cat.as_ordered()
        else:
            c_min = train_data[col].min()
            c_max = train_data[col].max()

            if pd.api.types.is_integer_dtype(train_data[col]):
                # Handle integer types
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    train_data[col] = train_data[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    train_data[col] = train_data[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    train_data[col] = train_data[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    train_data[col] = train_data[col].astype(np.int64)
            elif pd.api.types.is_float_dtype(train_data[col]):
                # Handle float types
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    train_data[col] = train_data[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    train_data[col] = train_data[col].astype(np.float32)
                else:
                    train_data[col] = train_data[col].astype(np.float64)

    end_mem = train_data.memory_usage().sum() / 1024**2
    print(f'Memory usage of dataframe is {start_mem:.2f} MB --> {end_mem:.2f} MB (Decreased by {100 * (start_mem - end_mem) / start_mem:.1f}%)')

    return train_data
```

### Importing data using dask for faster processing

```python
orders = dd.read_csv(root_csv + 'orders.csv')
departments = dd.read_csv(root_csv + 'departments.csv')
aisles = dd.read_csv(root_csv + 'aisles.csv')
products = dd.read_csv(root_csv + 'products.csv')
order_products_train = dd.read_csv(root_csv + 'order_products_train.csv')
order_products_prior = dd.read_csv(root_csv + 'order_products_prior.csv')
```

### Concatenating all dataframes into single unified dataframe for data analysis

```python
# Concatenate DataFrames along rows
order_products = dd.concat([order_products_prior, order_products_train], axis=0)

# Merge DataFrames using Dask
order_products = dd.merge(order_products, products, on='product_id', how='left')
order_products = dd.merge(order_products, aisles, on='aisle_id', how='left')
order_products = dd.merge(order_products, departments, on='department_id', how='left')
order_products = dd.merge(order_products, orders, on='order_id', how='left')

# Compute the result to obtain a Pandas DataFrame
order_products = order_products.compute()

# Creating a pickle file for unified datafrane
order_products.to_pickle(root_pickle + 'order_products.pkl')

# Reducing memory of unified dataframe
reduce_mem_usage(order_products)
```

## Aisles EDA

#### Creating dataframe specific to aisle for EDA

```python
aisle_EDA = order_products.groupby("aisle")["reordered"].agg(['count', 'sum']).rename(columns = {'count':'total','sum':'reorders'})
aisle_EDA = aisle_EDA.sort_values('total', ascending=False).reset_index()
aisle_EDA.head(5)
```

#### Total Orders and Reorders From Most Popular Aisles

```python
palette = sns.color_palette("Set2")
fig, ax = plt.subplots(figsize=(15, 8))
ax = sns.barplot(y=aisle_EDA.aisle[0:20], x=aisle_EDA.total[0:20], color=palette[0], label="total")
ax = sns.barplot(y=aisle_EDA.aisle[0:20], x=aisle_EDA.reorders[0:20], color=palette[3], label="reordered")
ax.set_ylabel("Aisle")
ax.set_xlabel("Orders Count")
ax.set_title("Total Orders and Reorders From Most Popular Aisles")
ax.legend(loc=4, prop={'size': 12})
plt.show()
```

#### Adding reorder ratio in aisle EDA dataframe

```python
aisle_EDA["reorder_ratio"] = aisle_EDA.reorders / aisle_EDA.total
aisle_EDA = aisle_EDA.sort_values("reorder_ratio", ascending=False).reset_index(drop=True)
```

#### Checking the highest reorder ratio for top 20 aisles

```python
palette = sns.color_palette("Set2")
fig, ax = plt.subplots(figsize=(13, 8))
ax = sns.barplot(y=aisle_EDA.aisle[0:20], x=aisle_EDA.reorder_ratio[0:20], color=palette[0])
ax.set_ylabel("Aisles")
ax.set_xlabel("Reorder Ratio")
ax.set_title("Aisles with Highest Reorder Ratio")
ax.tick_params(axis='both', labelsize=12)
plt.show()
```

## Department EDA

#### Creating dataframe specific to department for EDA

```python
dept_EDA = order_products.groupby("department")["reordered"].agg(['count', 'sum']).rename(columns = {'count':'total','sum':'reorders'})
dept_EDA = dept_EDA.sort_values('total', ascending=False).reset_index()
```

#### Total Orders and Reorders From Most Popular Departments

```python
fig, ax = plt.subplots(figsize = (15,8))
ax = sns.barplot(y = dept_EDA.department, x = dept_EDA["total"], color=palette[0], label = "total")
ax = sns.barplot(y = dept_EDA.department, x = dept_EDA["reorders"], color=palette[3], label = "reordered")
ax.set_ylabel("Department")
ax.set_xlabel("Frequency")
ax.legend(loc = 4, prop={'size': 12})
ax.set_title("Total Orders and Reorders From Departments")
plt.show()
```

#### Adding reorder ratio in department EDA dataframe

```python
dept_EDA["reorder_ratio"] = dept_EDA.reorders/dept_EDA.total
dept_EDA = dept_EDA.sort_values("reorder_ratio", ascending=False).reset_index()
```

#### Checking the top 20 departments with Highest Reorder Ratio

```python
fig, ax = plt.subplots(figsize = (13,8))
ax = sns.barplot(y = dept_EDA.department[0:20], x = dept_EDA.reorder_ratio, color=palette[0])
ax.set_ylabel("Departments")
ax.set_xlabel("Reorder Ratio")
ax.set_title("Departments with Highest Reorder Ratio")
ax.tick_params(axis = 'both', labelsize = 12)
plt.show()
```

## Product EDA

#### Creating dataframe specific to product for EDA

```python
prod_EDA = order_products.groupby("product_name")["reordered"].agg(['count', 'sum']).rename(columns = {'count':'total','sum':'reorders'})
prod_EDA = prod_EDA.sort_values('total', ascending=False).reset_index()
```

#### Total Orders and Reorders From Most Popular Products

```python
fig, ax = plt.subplots(figsize = (10,7))
ax = sns.barplot(y = prod_EDA.product_name[0:20], x = prod_EDA.total[0:20], color=palette[0], label = "total")
ax = sns.barplot(y = prod_EDA.product_name[0:20], x = prod_EDA.reorders[0:20], color=palette[3], label = "reordered")
ax.set_ylabel("Product")
ax.set_xlabel("Total Orders")
ax.set_title("Most Popular Products")
ax.legend(loc = 4, prop={'size': 12})
plt.show()
```

#### Adding reorder ratio in product EDA dataframe

```python
prod_EDA["reorder_ratio"] = prod_EDA.reorders/prod_EDA.total
prod_EDA.sort_values("reorder_ratio", ascending=False).head(5)
```

#### Creating a dataframe for total number of users ordered particular product

```python
product_unique_users_EDA = order_products.groupby('product_name')['user_id'].nunique().reset_index().rename(columns={'user_id':'total_users'})
product_unique_users_EDA.sort_values('total_users', ascending = False).head(5)
```

#### Product reorder ratio

```python
product_unique_users_EDA = product_unique_users_EDA.merge(prod_EDA, on='product_name', how='left')
product_unique_users_EDA.sort_values("reorder_ratio", ascending=False).head(5)
```

#### Adding organic column to check if the product is organic

```python
product_unique_users_EDA['Organic'] = product_unique_users_EDA.product_name.str.contains("Organic")
product_unique_users_EDA.head(5)
```

#### Organic Vs InOrganic

```python
fig, ax = plt.subplots(figsize = (5,5))
ax = sns.barplot(x = product_unique_users_EDA.groupby('Organic').size().index, y = product_unique_users_EDA.groupby('Organic').size().values)
ax.set_xlabel("Organic Product", size = 9)
ax.set_ylabel("Total Products", size = 9)
ax.set_title("Total Organic and Inorganic products", size = 10)
plt.show()
```

#### Reorder ratio for organic vs inorganic

```python
fig, ax = plt.subplots(figsize = (5,5))
ax = sns.barplot(x = product_unique_users_EDA.groupby('Organic')['reorder_ratio'].mean().index, y = product_unique_users_EDA.groupby('Organic')['reorder_ratio'].mean().values)
ax.set_xlabel("Organic Product", size = 9)
ax.set_ylabel("Mean reorder ratio", size = 9)
ax.set_title("Mean Reorder Ratio of Organic/Inorganic Products", size = 10)
plt.show()
```

#### Most Popular Products on different days

```python
prod_day_EDA = order_products.groupby(['order_dow', 'product_name']).size().reset_index(name='counts')
prod_day_EDA = prod_day_EDA.sort_values(['order_dow', 'counts'], ascending=[True, False])
prod_day_EDA = prod_day_EDA.groupby('order_dow').head(5).reset_index(drop = True)
ax = sns.catplot(x="order_dow", y="counts", hue="product_name", data=prod_day_EDA, kind="bar", legend=False)
ax.add_legend(title="Product")
ax.set_axis_labels("Day of Week", "Total Orders of Most Frequent Products")
ax.fig.suptitle("Most Popular Products on different Days of Week", va="baseline", ha="center")
ax.savefig("Most Popular Products on Different Days of Week.png")
```
